```python
def extract_hashtags(self):
    regex = "#(\w+)"
    hashtag_list = re.findall(regex, self)
    l=[]
    for x in hashtag_list:
        x = x.lower()
        x = re.sub(r"[^a-zA-Z ]+", '', x)
        l.append(x)
    return hashtag_list if hashtag_list else ''

def extract_tags(self):
    regex = "@(\w+)"
    name_tag = re.findall(regex, self)
    l=[]
    for x in name_tag:
        x = x.lower()
        x = re.sub(r"[^a-zA-Z ]+", '', x)
        l.append(x)
    return l if l else ''

def extract_links(self):
    regex = r'(https?://\S+)'
    url_list = re.findall(regex, self)
    l=[]
    for x in url_list:
        x = x.lower()
        x = re.sub(r"[^a-zA-Z ]+", '', x)
        l.append(x)
    return url_list if url_list else ''

def extract_emojis(self):
    emoji_list = ''.join(c for c in self if c in emoji.UNICODE_EMOJI['en'])
    text_emoji =  ''.join(emoji.demojize(x) for x in emoji_list if x)
    regex = r"\:(.*?)\:"
    text_emoji =  re.findall(regex, text_emoji)
    return text_emoji if text_emoji else ''

data['tweet_text'] = data['tweet_text'].str.lower()
data['hash_tag']   = data['tweet_text'].apply(extract_hashtags)
data['name_tag']   = data['tweet_text'].apply(extract_tags)
data['url_tag']    = data['tweet_text'].apply(extract_links)
data['emoji_tag']  = data['tweet_text'].apply(extract_emojis)
```

```python
data['tweet_text'] = data['tweet_text'].apply(lambda x: re.sub(r"[^a-zA-Z ]+", '', x))
temp = list(data.loc[data['tweet_text']==''].index)
```

```python
def clean_text(text,dirt):
    for x in dirt:
        text = text.replace(x,'')
        text = " ".join(text.split())
    return text
```

```python
# remove name_tag hash_tag and url_tag from tweets
data['tweet_text'] = data.apply(lambda row :clean_text(row['tweet_text'],row['name_tag']),axis=1)
temp_n = list(data.loc[data['tweet_text']==''].index)
data['tweet_text'] = data.apply(lambda row :clean_text(row['tweet_text'],row['hash_tag']),axis=1)
temp_h = list(data.loc[data['tweet_text']==''].index)
data['tweet_text'] = data.apply(lambda row :clean_text(row['tweet_text'],row['url_tag']),axis=1)
temp_u = list(data.loc[data['tweet_text']==''].index)
```
